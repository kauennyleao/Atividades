```python
def busca(array, valor):
    for i in range(len(array)):
        if valor == array[i]:
            return i  # Retorna índice do elemento encontrado
    return None


def insere(array, valor):
    posicao = busca(array, None)  # Busca posição vazia (None)
    if posicao != None:
        array[posicao] = valor
        return posicao
    else:
        print("Não inserido. Array completo?")
        return None


def remove(array, valor):
    posicao = busca(array, valor)  # Busca posição do valor

    if posicao == None:
        print("Valor não encontrado")
        return array

    for i in range(posicao, len(array) - 1):
        array[i] = array[i + 1]  # Realoca os elementos

    array[-1] = None  # Última posição vira vazia
    return array


def minimo(array):
    minimo = None
    for i in range(len(array)):
        if array[i] != None:
            if minimo == None or array[i] < minimo:
                minimo = array[i]
    return minimo


def maximo(array):
    maximo = None
    for i in range(len(array)):
        if array[i] != None:
            if maximo == None or array[i] > maximo:
                maximo = array[i]
    return maximo


def sucessor(array, valor):
    posicao = busca(array, valor)
    if posicao != None and posicao < len(array) - 1:
        return array[posicao + 1]
    return None


def predecessor(array, valor):
    posicao = busca(array, valor)
    if posicao != None and posicao > 0:
        return array[posicao - 1]
    return None


# 🔹 Teste do código
meuArray = [1, 2, 3, 4, 5, None, None, None]

print("Array inicial:", meuArray)

# Inserção
insere(meuArray, 6)

# Remoção
remove(meuArray, 3)

print("Array depois das operações:", meuArray)

# Busca
print("Posição do 5:", busca(meuArray, 5))

# Mínimo e máximo
print("Mínimo:", minimo(meuArray))
print("Máximo:", maximo(meuArray))

# Sucessor e predecessor
print("Sucessor de 4:", sucessor(meuArray, 4))
print("Predecessor de 4:", predecessor(meuArray, 4))
```
