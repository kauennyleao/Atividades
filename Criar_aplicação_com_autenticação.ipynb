```python
import getpass
import pickle
import hashlib
import os

def hash_senha(senha):
    return hashlib.sha256(senha.encode()).hexdigest()

def carregar():
    if os.path.exists("saved_data.pkl"):
        with open("saved_data.pkl", "rb") as f:
            return pickle.load(f)
    return {}

def salvar(dados):
    with open("saved_data.pkl", "wb") as f:
        pickle.dump(dados, f)

def cadastrar():
    user = input("Escolha o seu usuário: ")
    senha = getpass.getpass("Agora, escolha uma senha (Lembre-se dela): ")

    dados = carregar()
    dados[user] = hash_senha(senha)
    salvar(dados)

    print("Pronto! Agora você está cadastrado(a)!")

def login():
    user = input("Insira seu usuário: ")
    senha = getpass.getpass("Insira sua senha: ")

    dados = carregar()

    if user in dados and dados[user] == hash_senha(senha):
        print("Login válido!")
        return True
    else:
        print("Erro de login!")
        return False

def menu():
    print("\n--- MENU PRINCIPAL ---")
    print("1 - Continuar")
    print("2 - Sair")

# programa
print("1 - Cadastrar")
print("2 - Login")

op = input("Escolha uma das opções acima: ")

if op == "1":
    cadastrar()

elif op == "2":
    if login():   # 👈 verifica se deu certo
        menu()    # 👈 só entra aqui se login for válido
    else:
        print("Encerrando aplicação...")

else:
    print("Opção inválida!")
    print("Escolha 1 ou 2")
```
