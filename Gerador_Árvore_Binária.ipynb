```python

"""
🌳 GERADOR DE ÁRVORE BINÁRIA ALEATÓRIA
📚 Estrutura de Dados - Exemplo Simples
"""

import random


class NoArvore:
    # Classe do nó da árvore
    def __init__(self, isbn):
        self.isbn = isbn
        self.esquerda = None
        self.direita = None


class ArvoreBinaria:
    # Classe da árvore binária
    def __init__(self):
        self.raiz = None

    # Função para inserir valores
    def inserir(self, isbn):
        if self.raiz is None:
            self.raiz = NoArvore(isbn)
        else:
            self._inserir_recursivo(self.raiz, isbn)

    # Inserção recursiva
    def _inserir_recursivo(self, no, isbn):
        if isbn < no.isbn:
            if no.esquerda is None:
                no.esquerda = NoArvore(isbn)
            else:
                self._inserir_recursivo(no.esquerda, isbn)
        else:
            if no.direita is None:
                no.direita = NoArvore(isbn)
            else:
                self._inserir_recursivo(no.direita, isbn)

    # Função para imprimir a árvore
    def imprimir_arvore(self):
        if self.raiz is None:
            print("Árvore vazia!")
            return

        print("\n🌳 PRONTO! SUA ÁRVORE BINÁRIA GERADA:")
        self._imprimir_recursivo(self.raiz, "", True)

    # Impressão recursiva
    def _imprimir_recursivo(self, no, prefixo, eh_direita):
        if no is None:
            return

        print(prefixo, end="")

        if eh_direita:
            print("└── ", end="")
            novo_prefixo = prefixo + "    "
        else:
            print("├── ", end="")
            novo_prefixo = prefixo + "│   "

        print(no.isbn)

        if no.esquerda:
            self._imprimir_recursivo(no.esquerda, novo_prefixo, False)

        if no.direita:
            self._imprimir_recursivo(no.direita, novo_prefixo, True)


# PROGRAMA PRINCIPAL
def main():

    print("=" * 50)
    print("🌳 GERADOR DE ÁRVORE BINÁRIA")
    print("=" * 50)

    # Solicita quantidade de nós
    quantidade = int(input("Olá! Digite a quantidade de nós da sua árvore: "))

    # Cria a árvore
    arvore = ArvoreBinaria()

    print("\nNúmeros gerados aleatoriamente:")

    # Gera números aleatórios e insere na árvore
    for i in range(quantidade):
        numero = random.randint(1, 100)

        print(numero)

        arvore.inserir(numero)

    # Imprime a árvore
    arvore.imprimir_arvore()


# Executa o programa
if __name__ == "__main__":
    main()
```
